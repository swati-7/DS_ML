### Observation
- Married Applicants have a higher probability of getting a Loan
- use a IF then statement to Predict the "y"/dependent variable
- Use apply(), lambda function on your selected categorical features

```python
#CodeSection14
train['Prediction']=train['relevent_experience'].apply(lambda x : 1 if x == 'Has relevent experience' else 0)
```

### We can use a Accuracy Function from Metrics

- Within it we will choose the "accuracy_score" function

```python
#CodeSection16
from sklearn.metrics import accuracy_score
```

```python
#CodeSection17
y_actual=train['target']
y_predict=train['Prediction']
```

```python
#CodeSection18
print(f" Train Accuracy : {accuracy_score(y_actual,y_predict):0.1%}")
```

## 4/ Impute Categorical Variables
- Impute all Numerical Variables

### Categorical Variable/Feature 1

```python
#CodeSection9
# Impute Feature with mode (use [0]) for both train and test - use for loop
for dataset in all_data:
  dataset['gender']=dataset['gender'].fillna(train['gender'].mode()[0])
```

### Categorical Variable/Feature 2

```python
#CodeSection9
# Impute Feature with mode (use [0]) for both train and test - use for loop
#gender	relevent_experience	enrolled_university	education_level	major_discipline	experience	company_size	company_type	last_new_job
for dataset in all_data:
  dataset['relevent_experience']=dataset['relevent_experience'].fillna(train['relevent_experience'].mode()[0])
```

### Categorical Variable/Feature 3

```python
#CodeSection9
# Impute Feature with mode (use [0]) for both train and test - use for loop
for dataset in all_data:
  dataset['enrolled_university']=dataset['enrolled_university'].fillna(train['enrolled_university'].mode()[0])
```

### Categorical Variable/Feature 4

```python
#CodeSection9
# Impute Feature with mode (use [0]) for both train and test - use for loop
for dataset in all_data:
  dataset['education_level']=dataset['education_level'].fillna(train['education_level'].mode()[0])
  dataset['major_discipline']=dataset['major_discipline'].fillna(train['major_discipline'].mode()[0])
  dataset['experience']=dataset['experience'].fillna(train['experience'].mode()[0])
  dataset['company_size']=dataset['company_size'].fillna(train['company_size'].mode()[0])
  dataset['company_type']=dataset['company_type'].fillna(train['company_type'].mode()[0])
  dataset['last_new_job']=dataset['last_new_job'].fillna(train['last_new_job'].mode()[0])
```

## 6/ Prepare Data for Model Model

```python
#CodeSection11
# Check first 5 rows of Test Dataset
test.head()
```

```python
#CodeSection12
# Divide Data into X and y
X=train.drop(['enrollee_id','target'],axis=1)
y=train['target']
```

### Create Dummy Variables for  Categorical Variables

```python
#CodeSection13
# Create Dummy Variables
X=pd.get_dummies(X,drop_first=True)
```

```python
X.shape
```

```python
#CodeSection14
# Check first 5 rows of X
X.head()
```

## 7/ Build Model

```python
#CodeSection15
#Start by importing LogisticRegression 
from sklearn.linear_model import LogisticRegression

#Initialize LogisticRegression
clf = LogisticRegression(max_iter=1000)
```

```python
#CodeSection16
# Fit Model
model = clf.fit(X,y)
```

### Check Accuracy of Model on Train Data

```python
#CodeSection17
# Predict on Train Data and check predictions
y_pred = model.predict(X)
y_pred
```
